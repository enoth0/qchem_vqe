```python
import numpy as np
import matplotlib.pyplot as plt

from qiskit_nature.second_q.drivers import PySCFDriver
from qiskit_nature.units import DistanceUnit
from qiskit_nature.second_q.mappers import ParityMapper
from qiskit_nature.second_q.circuit.library import UCCSD, HartreeFock

# GPU-enabled imports
from qiskit_aer import AerSimulator
from qiskit.primitives import BackendEstimator
from qiskit_algorithms import VQE
from qiskit_algorithms.optimizers import SLSQP, COBYLA

from pyscf import gto, scf, fci

def get_triangle_coords(d):
    """Returns coordinates for H3+ in equilateral triangle arrangement"""
    h = d * np.sqrt(3) / 2
    return [[0,0,0], [d,0,0], [d/2,h,0]]

def get_pyscf_mol(coords, charge, spin, basis):
    atom_spec = ""
    for i, c in enumerate(coords):
        atom_spec += f"H {c[0]} {c[1]} {c[2]}; "
    atom_spec = atom_spec.strip("; ")
    mol = gto.Mole()
    mol.atom = atom_spec
    mol.basis = basis
    mol.charge = charge
    mol.spin = spin
    mol.build()
    return mol

def get_qiskit_problem(coords, charge, spin, basis):
    atom_spec = ""
    for c in coords:
        atom_spec += f"H {c[0]} {c[1]} {c[2]}; "
    atom_spec = atom_spec.strip("; ")
    driver = PySCFDriver(atom=atom_spec, unit=DistanceUnit.ANGSTROM, charge=charge, spin=spin, basis=basis)
    return driver.run()

def run_uhf_pyscf(mol):
    mf = scf.UHF(mol)
    ehf = mf.kernel()
    return ehf

def run_fci_pyscf(mol):
    mf = scf.RHF(mol).run()
    cisolver = fci.FCI(mol, mf.mo_coeff)
    efci, _ = cisolver.kernel()
    return efci

def run_vqe_gpu(problem, mapper):
    """Modified VQE function to use GPU acceleration"""
    
    # Create GPU-enabled simulator
    gpu_simulator = AerSimulator(
        method='statevector',  # Use statevector method for better GPU performance
        device='GPU',          # Specify GPU device
        precision='single'     # Use single precision for faster computation
    )
    
    # Set GPU-specific options for better performance
    gpu_simulator.set_options(
        max_parallel_threads=0,      # Use all available GPU threads
        statevector_parallel_threshold=8,  # Threshold for parallel execution
        blocking_enable=True,        # Enable blocking for better memory management
        blocking_qubits=10          # Number of qubits per block
    )
    
    print(f"Using backend: {gpu_simulator.name}")
    print(f"Backend configuration: {gpu_simulator.configuration().to_dict()}")
    
    # Use BackendEstimator with GPU simulator
    estimator = BackendEstimator(backend=gpu_simulator)
    
    optimizer = COBYLA(maxiter=200)
    init_state = HartreeFock(problem.num_spatial_orbitals, problem.num_particles, mapper)
    ansatz = UCCSD(problem.num_spatial_orbitals, problem.num_particles, mapper, initial_state=init_state)
    
    print(f"Number of qubits: {ansatz.num_qubits}")
    print(f"Number of parameters: {ansatz.num_parameters}")
    
    vqe = VQE(estimator, ansatz, optimizer, initial_point=[0]*ansatz.num_parameters)
    qubit_op = mapper.map(problem.hamiltonian.second_q_op())
    
    print(f"Starting VQE optimization...")
    result = vqe.compute_minimum_eigenvalue(qubit_op)
    
    # IMPORTANT: Use interpret to get total energy!
    energy = problem.interpret(result).total_energies[0].real
    return energy

def check_gpu_availability():
    """Check if GPU is available for Aer simulator"""
    try:
        gpu_simulator = AerSimulator(device='GPU')
        print(f"GPU available: {gpu_simulator.available_devices()}")
        return True
    except Exception as e:
        print(f"GPU not available: {e}")
        print("Falling back to CPU simulation")
        return False

# --- Main Calculation Setup ---
print("Checking GPU availability...")
gpu_available = check_gpu_availability()

basis = 'sto3g'
charge = 1
spin = 0
dists = np.linspace(0.4, 4.0, 14)

vqe_energies, uhf_energies, fci_energies = [], [], []

# Loop through different H-H distances
for i, d in enumerate(dists):
    print(f"\nCalculation {i+1}/{len(dists)}: d = {d:.2f} Ã…")
    coords = get_triangle_coords(d)
    
    # VQE with GPU
    try:
        problem = get_qiskit_problem(coords, charge, spin, basis)
        mapper = ParityMapper(num_particles=problem.num_particles)
        
        if gpu_available:
            vqe_energy = run_vqe_gpu(problem, mapper)
        else:
            # Fallback to CPU if GPU not available
            from qiskit.primitives import Estimator
            estimator = Estimator()
            optimizer = COBYLA(maxiter=200)
            init_state = HartreeFock(problem.num_spatial_orbitals, problem.num_particles, mapper)
            ansatz = UCCSD(problem.num_spatial_orbitals, problem.num_particles, mapper, initial_state=init_state)
            vqe = VQE(estimator, ansatz, optimizer, initial_point=[0]*ansatz.num_parameters)
            qubit_op = mapper.map(problem.hamiltonian.second_q_op())
            result = vqe.compute_minimum_eigenvalue(qubit_op)
            vqe_energy = problem.interpret(result).total_energies[0].real
            
    except Exception as e:
        vqe_energy = np.nan
        print(f"VQE failed at d={d:.2f}: {e}")

    # UHF (CPU-based, no GPU acceleration available)
    try:
        mol = get_pyscf_mol(coords, charge, spin, basis)
        uhf_energy = run_uhf_pyscf(mol)
    except Exception as e:
        uhf_energy = np.nan
        print(f"UHF failed at d={d:.2f}: {e}")

    # FCI (CPU-based, no GPU acceleration available)
    try:
        fci_energy = run_fci_pyscf(mol)
    except Exception as e:
        fci_energy = np.nan
        print(f"FCI failed at d={d:.2f}: {e}")

    vqe_energies.append(vqe_energy)
    uhf_energies.append(uhf_energy)
    fci_energies.append(fci_energy)
    print(f"Results: VQE={vqe_energy:.6f} | UHF={uhf_energy:.6f} | FCI={fci_energy:.6f}")

# Plot potential energy curves
plt.figure(figsize=(10, 6))
plt.plot(dists, vqe_energies, 'o-', label='VQE (GPU)', color='limegreen', linewidth=2)
plt.plot(dists, uhf_energies, 's-', label='UHF (CPU)', color='red', linewidth=2)
plt.plot(dists, fci_energies, '--', label='FCI (CPU)', color='black', linewidth=2)
plt.xlabel("H-H Distance (Ã…)", fontsize=12)
plt.ylabel("Energy (Hartree)", fontsize=12)
plt.title("H$_3^+$ Potential Energy Curve (GPU-Accelerated VQE)", fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Performance comparison
print(f"\n{'='*50}")
print("PERFORMANCE SUMMARY")
print(f"{'='*50}")
print(f"GPU Available: {gpu_available}")
if gpu_available:
    print("VQE calculations performed on GPU")
else:
    print("VQE calculations performed on CPU (fallback)")
print(f"Total calculations completed: {len([e for e in vqe_energies if not np.isnan(e)])}/{len(dists)}")
```

```python
# Find minimum VQE energy (ground state energy)
min_vqe_index = vqe_energies.index(min(vqe_energies))
min_vqe_energy = vqe_energies[min_vqe_index]
min_vqe_distance = dists[min_vqe_index]

print(f"H3+ Ground State Energy (VQE):")
print(f"Energy: {min_vqe_energy:.6f} Hartree")
print(f"At bond distance: {min_vqe_distance:.3f} Ã…")

# For comparison, also show FCI result
min_fci_index = fci_energies.index(min(fci_energies))
min_fci_energy = fci_energies[min_fci_index]
min_fci_distance = dists[min_fci_index]

print(f"\nH3+ Ground State Energy (FCI/Exact):")
print(f"Energy: {min_fci_energy:.6f} Hartree")
print(f"At bond distance: {min_fci_distance:.3f} Ã…")

print(f"\nError: {(min_vqe_energy - min_fci_energy)*627.5:.2f} kcal/mol")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Core Qiskit imports
from qiskit_aer import AerSimulator
from qiskit.primitives import BackendEstimator
from qiskit_algorithms import VQE
from qiskit_algorithms.optimizers import SLSQP

# Qiskit Nature imports
from qiskit_nature.second_q.drivers import PySCFDriver
from qiskit_nature.units import DistanceUnit
from qiskit_nature.second_q.mappers import ParityMapper
from qiskit_nature.second_q.circuit.library import HartreeFock, UCCSD

# PySCF for classical reference
from pyscf import gto, scf

def setup_gpu_backend():
    """Setup GPU-accelerated Aer simulator for H2 calculations"""
    
    print("Setting up GPU backend...")
    
    try:
        # Create GPU simulator
        gpu_simulator = AerSimulator(
            method='statevector',  # Exact simulation for VQE
            device='GPU',          # Use GPU acceleration
            precision='single'     # Single precision for speed
        )
        
        # GPU optimization settings for H2 (small molecule)
        gpu_simulator.set_options(
            max_parallel_threads=0,           # Use all GPU cores
            statevector_parallel_threshold=4, # Low threshold for H2
            blocking_enable=True,             # Memory blocking
            blocking_qubits=4,                # Small blocks for H2
            batched_optimization=True         # Batch operations
        )
        
        # Test GPU availability
        available_devices = gpu_simulator.available_devices()
        print(f"Available devices: {available_devices}")
        
        if 'GPU' in str(available_devices):
            print("âœ“ GPU backend successfully initialized")
            return gpu_simulator, True
        else:
            raise Exception("No GPU devices detected")
            
    except Exception as e:
        print(f"âœ— GPU setup failed: {e}")
        print("Please check your qiskit-aer-gpu-cu11 installation")
        return None, False

def create_h2_problem(distance, basis='sto-3g'):
    """Create H2 molecular problem using PySCF driver"""
    
    # H2 molecule: two hydrogen atoms along z-axis
    atom_string = f"H 0 0 0; H 0 0 {distance}"
    
    driver = PySCFDriver(
        atom=atom_string,
        unit=DistanceUnit.ANGSTROM,
        charge=0,  # Neutral H2
        spin=0,    # Singlet state
        basis=basis
    )
    
    return driver.run()

def calculate_classical_reference(distance, basis='sto-3g'):
    """Calculate classical HF energy using PySCF"""
    
    mol = gto.Mole()
    mol.atom = f"H 0 0 0; H 0 0 {distance}"
    mol.basis = basis
    mol.charge = 0
    mol.spin = 0
    mol.verbose = 0
    mol.build()
    
    # Hartree-Fock calculation
    mf = scf.RHF(mol)
    mf.verbose = 0
    hf_energy = mf.kernel()
    
    return hf_energy

def run_h2_vqe_gpu(distance, gpu_backend, basis='sto-3g'):
    """Run VQE calculation for H2 on GPU"""
    
    print(f"\nCalculating H2 at {distance:.2f} Ã… using GPU VQE...")
    
    # Create molecular problem
    problem = create_h2_problem(distance, basis)
    print(f"H2 electrons: {problem.num_particles}")
    print(f"H2 spatial orbitals: {problem.num_spatial_orbitals}")
    
    # Map to qubits using parity mapping (efficient for H2)
    mapper = ParityMapper(num_particles=problem.num_particles)
    
    # Get Hamiltonian in qubit form
    hamiltonian = problem.hamiltonian
    qubit_op = mapper.map(hamiltonian.second_q_op())
    print(f"Qubits needed: {qubit_op.num_qubits}")
    print(f"Hamiltonian terms: {len(qubit_op)}")
    
    # Create GPU estimator
    estimator = BackendEstimator(backend=gpu_backend)
    
    # Create initial state (Hartree-Fock)
    initial_state = HartreeFock(
        num_spatial_orbitals=problem.num_spatial_orbitals,
        num_particles=problem.num_particles,
        qubit_mapper=mapper
    )
    
    # Create UCCSD ansatz (gold standard for molecules)
    ansatz = UCCSD(
        num_spatial_orbitals=problem.num_spatial_orbitals,
        num_particles=problem.num_particles,
        qubit_mapper=mapper,
        initial_state=initial_state
    )
    print(f"Ansatz parameters: {ansatz.num_parameters}")
    
    # Setup VQE with GPU
    optimizer = SLSQP(maxiter=200, ftol=1e-8)
    vqe = VQE(estimator, ansatz, optimizer)
    
    # Run VQE on GPU
    start_time = time.time()
    result = vqe.compute_minimum_eigenvalue(qubit_op)
    gpu_time = time.time() - start_time
    
    # Extract total energy
    total_energy = problem.interpret(result).total_energies[0].real
    
    print(f"âœ“ GPU VQE completed in {gpu_time:.3f} seconds")
    print(f"Ground state energy: {total_energy:.8f} Hartree")
    print(f"Optimizer evaluations: {result.cost_function_evals}")
    
    return total_energy, gpu_time, result.cost_function_evals

def main():
    """Main H2 VQE calculation with GPU acceleration"""
   
    
    # Setup GPU backend
    gpu_backend, gpu_available = setup_gpu_backend()
    
    if not gpu_available:
        print("GPU not available. Exiting...")
        return
    
    # H2 bond distances to calculate (around equilibrium ~0.74 Ã…)
    distances = np.linspace(0.5, 2.0, 8)
    
    # Storage for results
    vqe_energies = []
    hf_energies = []
    gpu_times = []
    total_evaluations = []
    
    print(f"\nCalculating H2 potential energy surface...")
    print(f"Bond distances: {distances[0]:.2f} - {distances[-1]:.2f} Ã…")
    print("-" * 60)
    
    for i, d in enumerate(distances):
        print(f"\nStep {i+1}/{len(distances)}: H-H distance = {d:.2f} Ã…")
        
        try:
            # VQE calculation on GPU
            vqe_energy, calc_time, evaluations = run_h2_vqe_gpu(d, gpu_backend)
            vqe_energies.append(vqe_energy)
            gpu_times.append(calc_time)
            total_evaluations.append(evaluations)
            
            # Classical HF reference
            hf_energy = calculate_classical_reference(d)
            hf_energies.append(hf_energy)
            
            # Results comparison
            correlation_energy = vqe_energy - hf_energy
            print(f"HF energy:     {hf_energy:.8f} Hartree")
            print(f"VQE energy:    {vqe_energy:.8f} Hartree") 
            print(f"Correlation:   {correlation_energy:.8f} Hartree")
            print(f"Correlation:   {correlation_energy*627.5:.4f} kcal/mol")
            
        except Exception as e:
            print(f"âœ— Calculation failed: {e}")
            vqe_energies.append(np.nan)
            hf_energies.append(np.nan)
            gpu_times.append(np.nan)
            total_evaluations.append(0)
    
    # Analysis
    print("\n" + "=" * 60)
    print("RESULTS ANALYSIS")
    print("=" * 60)
    
    # Performance summary
    successful_calcs = len([e for e in vqe_energies if not np.isnan(e)])
    total_gpu_time = sum(t for t in gpu_times if not np.isnan(t))
    total_evals = sum(total_evaluations)
    
    print(f"Successful calculations: {successful_calcs}/{len(distances)}")
    print(f"Total GPU time: {total_gpu_time:.2f} seconds")
    print(f"Average time per point: {total_gpu_time/successful_calcs:.2f} seconds")
    print(f"Total VQE evaluations: {total_evals}")
    
    # Find equilibrium
    if successful_calcs > 0:
        valid_indices = [i for i, e in enumerate(vqe_energies) if not np.isnan(e)]
        min_energy_idx = min(valid_indices, key=lambda i: vqe_energies[i])
        equilibrium_distance = distances[min_energy_idx]
        equilibrium_energy = vqe_energies[min_energy_idx]
        
        print(f"\nEquilibrium geometry (VQE):")
        print(f"Bond length: {equilibrium_distance:.3f} Ã…")
        print(f"Ground state energy: {equilibrium_energy:.8f} Hartree")
        print(f"Experimental H2 bond length: 0.741 Ã…")
    
    # Plotting
    plt.figure(figsize=(15, 10))
    
    # Main potential energy curve
    plt.subplot(2, 2, 1)
    plt.plot(distances, vqe_energies, 'o-', label='VQE (GPU)', color='green', linewidth=3, markersize=8)
    plt.plot(distances, hf_energies, 's-', label='Hartree-Fock', color='red', linewidth=2, markersize=6)
    plt.axvline(x=0.741, color='gray', linestyle='--', alpha=0.7, label='Experimental (0.741 Ã…)')
    plt.xlabel('H-H Distance (Ã…)', fontsize=12)
    plt.ylabel('Energy (Hartree)', fontsize=12)
    plt.title('Hâ‚‚ Potential Energy Surface\n(GPU-Accelerated VQE)', fontsize=14)
    plt.legend(fontsize=11)
    plt.grid(True, alpha=0.3)
    
    # Correlation energy
    plt.subplot(2, 2, 2)
    correlation_energies = [(v - h)*627.5 if not (np.isnan(v) or np.isnan(h)) else np.nan 
                           for v, h in zip(vqe_energies, hf_energies)]
    plt.plot(distances, correlation_energies, 'o-', color='purple', linewidth=2, markersize=6)
    plt.xlabel('H-H Distance (Ã…)', fontsize=12)
    plt.ylabel('Correlation Energy (kcal/mol)', fontsize=12)
    plt.title('Electron Correlation Energy', fontsize=12)
    plt.grid(True, alpha=0.3)
    
    # GPU performance
    plt.subplot(2, 2, 3)
    plt.plot(distances, gpu_times, 'o-', color='orange', linewidth=2, markersize=6)
    plt.xlabel('H-H Distance (Ã…)', fontsize=12)
    plt.ylabel('GPU Time (seconds)', fontsize=12)
    plt.title('GPU Performance', fontsize=12)
    plt.grid(True, alpha=0.3)
    
    # Optimizer evaluations
    plt.subplot(2, 2, 4)
    plt.plot(distances, total_evaluations, 'o-', color='brown', linewidth=2, markersize=6)
    plt.xlabel('H-H Distance (Ã…)', fontsize=12)
    plt.ylabel('VQE Function Evaluations', fontsize=12)
    plt.title('Optimizer Convergence', fontsize=12)
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nâœ“ H2 VQE calculation completed successfully using GPU acceleration!")
    
    return {
        'distances': distances,
        'vqe_energies': vqe_energies,
        'hf_energies': hf_energies,
        'gpu_times': gpu_times,
        'evaluations': total_evaluations
    }

if __name__ == "__main__":
    results = main()
```
